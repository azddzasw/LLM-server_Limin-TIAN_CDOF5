# **Convert an Excel file (xlsx) to a CSV file**

```python
import pandas as pd

# Replace 'input.xlsx' with the path to your Excel file
xlsx_file = '/content/User stories for xlsx_V1.xlsx'

# Replace 'output.csv' with the desired path and filename for the CSV file
csv_file = 'output.csv'

try:
    # Read Excel file into a pandas DataFrame
    df = pd.read_excel(xlsx_file)

    # Write DataFrame to a CSV file
    df.to_csv(csv_file, index=False)

    print(f"Conversion complete. CSV file saved to {csv_file}")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import re
from fuzzywuzzy import fuzz
import pandas as pd

# Read Excel file into a pandas DataFrame
#xlsx_file = '/content/9_user_stories_modified.xlsx'
#df = pd.read_excel(xlsx_file)

# Define keywords
keywords = ['value statement', 'Description', 'user stories', 'User stories']

# Build the regular expression pattern
keyword_pattern = re.compile(r'(' + '|'.join(map(re.escape, keywords)) + ')', flags=re.IGNORECASE)

# Similarity matching threshold
threshold = 60  # Adjust the threshold as needed

# Find columns with similarity to the target
similar_columns = [col for col in df.columns if fuzz.partial_ratio(col.lower(), 'user story') > threshold]

if similar_columns:
    # Choose the best match based on similarity
    best_match = max(similar_columns, key=lambda col: fuzz.partial_ratio(col.lower(), 'user story'))

    # Use regular expression to match column names
    matching_columns = [col for col in df.columns if keyword_pattern.search(col)]

    # Output results
    print("Similarity Matching Results:", similar_columns)
    print("Best Match:", best_match)
    print("Regular Expression Matching Results:", matching_columns)
else:
    print("No columns found with sufficient similarity.")
```
